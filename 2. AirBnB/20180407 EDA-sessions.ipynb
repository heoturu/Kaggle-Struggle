```python
# f_act = df_sessions.action.value_counts().argsort() 
# f_act_detail = df_sessions.action_detail.value_counts().argsort()
f_act_type = df_sessions.action_type.value_counts().argsort() 
# f_dev_type = df_sessions.device_type.value_counts().argsort() 
f_new_act_detail = df_sessions.new_action_detail.value_counts().argsort() 
```

```python
dgr_sess = df_sessions.groupby(['user_id'])
```

```python
samples = []
cont = 0
ln = len(dgr_sess)

for g in dgr_sess:
    if cont%10000 == 0:
        print("%s from %s" %(cont, ln))
    # 아이디 당 세션 정보
    gr = g[1]
    l = []
    
    # 해당 id 
    l.append(g[0])
    

#   new_action_detail features

    c_new_act_detail = [0] * len(f_new_act_detail)
    for index, value in zip(gr.new_action_detail.index, gr.new_action_detail.values):  
        # one hot vector
        # i: value count, v: name of action
        # f_act = action unique, count 
        # one hot vector: c_act 해당 클래스 부분에 1로 표시     
        c_new_act_detail[f_new_act_detail[value]] += gr.loc[index, 'secs_elapsed']

          
    l = l + c_new_act_detail
    
    
    #action_type features
    c_act_type = [0] * len(f_act_type)
    for i,v in enumerate(gr.action_type.values):
        c_act_type[f_act_type[v]] += 1  
        
    _, c_act_type_uqc = np.unique(gr.action_type.values, return_counts=True)
    l = l + c_act_type       

    # 해당 id의 세션 row 개수 (= 해당 id 의 action 개수) 
    l.append(len(gr))
    sev = gr.secs_elapsed.fillna(0).values   #These values are used later.
    
    samples.append(l)
    cont += 1

# 결과적으로 for 문을 통해서 아이디 당 sessions.csv 의 action, action_type(count, log), action_detail, device_type, 
# secs_elapsed(log - sum, mean, std, median, l_secs[0] / float(l[1])) 더미 변수로 변경
```

```python
col_names = []    #name of the columns


for i in range(len(f_new_act_detail)):
    col_names.append(f_new_act_detail[f_new_act_detail.values == i].index[0])
len(col_names)

#Creating a dataframe with the computed features    
for i in range(len(f_act_type)):
    col_names.append('act_type' + str(i)) 
len(col_names)
                     
col_names.append('act_count')
```

```python
#preparing objects    
samples = np.array(samples)
samp_ar = samples[:, 1:]
samp_id = samples[:, 0]   #The first element in obs is the id of the sample.
```

```python
#creating the dataframe        
df_agg_sess = pd.DataFrame(samp_ar, columns=col_names)
df_agg_sess['id'] = samp_id 
df_agg_sess.index = samp_id 
# df_agg_sess.drop("id", axis=1, inplace=True)
```
